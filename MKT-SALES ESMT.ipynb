```python
# Scale the data for factor analysis
scaler = StandardScaler()
df_scaled = scaler.fit_transform(data)
```

```python
# Perform factor analysis
factor_analysis = FactorAnalysis(n_components=df_scaled.shape[1], random_state=42)
factor_scores = factor_analysis.fit_transform(df_scaled)

# Calculate eigenvalues
cov_matrix = factor_analysis.get_covariance()
eigenvalues, _ = np.linalg.eig(cov_matrix)
eigenvalues = np.real(eigenvalues)

# Sort eigenvalues in descending order
eigenvalues_sorted = np.sort(eigenvalues)[::-1]

# Calculate explained variance ratio
explained_variance_ratio = eigenvalues_sorted / eigenvalues_sorted.sum()
cumulative_variance = np.cumsum(explained_variance_ratio)

# Plot scree plot
plt.plot(range(1, df_scaled.shape[1] + 1), cumulative_variance, marker='o', linestyle='-')
plt.axvline(x=5, color='r', linestyle='--')  # Add a vertical line at 5 factors
plt.xlabel('Number of Factors')
plt.ylabel('Cumulative Variance Explained')
plt.title('Scree Plot')
plt.xticks(range(1, df_scaled.shape[1] + 1))
plt.show()
```

```python
import numpy as np
import pandas as pd
from factor_analyzer import FactorAnalyzer

# Convert df_scaled to DataFrame
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)

# Calculate eigenvalues of the correlation matrix
correlation_matrix = np.corrcoef(df_scaled.T)
eigenvalues, _ = np.linalg.eig(correlation_matrix)

# Sort eigenvalues in descending order
sorted_eigenvalues = np.sort(eigenvalues)[::-1]

# Calculate cumulative variance explained
cumulative_variance = np.cumsum(sorted_eigenvalues) / np.sum(sorted_eigenvalues)

# Print the cumulative variance explained by each factor
print("Cumulative Variance Explained:")
for i, variance in enumerate(cumulative_variance[:5]):
    factor_name = f"Factor {i+1}"
    print(f"{factor_name}: {variance:.2%}")

# Perform factor analysis
n_factors = 5
factor_analyzer = FactorAnalyzer(n_factors=n_factors, rotation=None)
factor_analyzer.fit(df_scaled)

# Get the factor loadings from the factor analyzer object
factor_loadings = factor_analyzer.loadings_

# Number of top loading variables to consider
top_loading_vars = 3

# Get the indices of variables with highest loadings for each factor
top_loading_indices = np.argsort(np.abs(factor_loadings), axis=0)[::-1][:top_loading_vars]

# Get the names of the variables from your original data
variable_names = list(df_scaled.columns)

# Print the variables with highest loadings for each factor
print("\nFactor Loadings:")
for i, factor_index in enumerate(top_loading_indices.T):
    factor_name = f"Factor {i+1}"
    variable_names_selected = [variable_names[j] for j in factor_index]
    print(f"{factor_name}: {variable_names_selected}")
```
